```python
import pandas as pd
import os
import numpy as np
```

```python
REF_SINGER = "MCUR"

features = ["melodiousness","articulation","rhythmic_complexity","rhythm_stability","dissonance","tonality","modality"]
df = pd.DataFrame(columns =features)
for dir,subdir,files in os.walk('./'):
    for file in files:
        path = os.path.join(dir,file)
        if ".npy" in file and "MCUR" not in file:
            dataline = np.load(path)   
            dataline = pd.DataFrame(dataline,columns=features )
            df = df.append(dataline,ignore_index=True)
            
```
