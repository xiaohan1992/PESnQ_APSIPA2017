```python
import pandas as pd
import os
import numpy as np
```

```python
REF_SINGER = "MCUR"

features = ["melodiousness","articulation","rhythmic_complexity","rhythm_stability","dissonance","tonality","modality","song_name"]
df_7_features = pd.DataFrame(columns =features)
for dir,subdir,files in os.walk('./'):
    for file in files:
        path = os.path.join(dir,file)
        if ".npy" in file and "MCUR" not in file:
            name = file.split('.')[0]
            dataline = list(np.load(path) )
            dataline.append(name)
            new = {}
            for i,data in enumerate(dataline):
                new[features[i]]=data
            df_7_features = df_7_features.append(new,ignore_index=True)
            
            
            
```
